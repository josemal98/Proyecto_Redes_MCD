### Cálculos Renales

<p align="justify">
Los cálculos renales, también conocidos como litiasis renal o nefrolitiasis, son una afección común que afecta a un gran número de personas en todo el mundo. Estos son pequeños depósitos sólidos que se forman en los riñones a partir de sustancias presentes en la orina. Los cálculos renales pueden variar en tamaño, desde diminutos granos de arena hasta piedras más grandes que pueden obstruir los conductos urinarios.

<p align="justify">
Existen diferentes tipos de cálculos renales, y su composición puede estar compuesta principalmente de sales de calcio, ácido úrico, estruvita o cistina. Las causas exactas de su formación pueden variar, pero en general, los factores de riesgo incluyen una combinación de desequilibrios químicos en la orina, baja ingesta de líquidos, dieta poco saludable, obesidad, historial familiar de cálculos renales y ciertas condiciones médicas como la hiperparatiroidismo o la enfermedad inflamatoria intestinal.

<p align="justify">
Los efectos negativos de los cálculos renales en la salud pueden ser significativos. Estos pueden causar dolor intenso en el costado o la espalda baja, así como síntomas como náuseas, vómitos, sangre en la orina e infecciones recurrentes del tracto urinario. Además, los cálculos renales pueden bloquear el flujo normal de orina, lo que puede llevar a complicaciones graves, como daño renal o infecciones graves.

<p align="justify">
El diagnóstico de los cálculos renales se realiza generalmente mediante una combinación de síntomas clínicos, análisis de orina, pruebas de imagen y, en algunos casos, pruebas de laboratorio adicionales. Sin embargo, los métodos de diagnóstico convencionales pueden presentar limitaciones. Por ejemplo, los análisis de orina actuales pueden ser subjetivos y dependen de la interpretación del especialista. Además, las pruebas de imagen pueden no detectar cálculos pequeños o pueden requerir radiación ionizante, lo que plantea preocupaciones en términos de exposición y seguridad.

<p align="justify">
En este contexto, las redes neuronales, un enfoque de inteligencia artificial inspirado en el funcionamiento del cerebro humano, han surgido como una herramienta prometedora en el diagnóstico de cálculos renales. Estas redes neuronales pueden aprender patrones complejos a partir de grandes conjuntos de datos, lo que permite desarrollar modelos capaces de detectar y clasificar cálculos renales en base al análisis de orina.

<p align="justify">
Las estadísticas revelan que los cálculos renales son una afección común en todo el mundo, afectando aproximadamente al 10% de la población en algún momento de su vida. Además, la incidencia de cálculos renales parece estar en aumento, posiblemente debido a cambios en la dieta y los estilos de vida.

<p align="justify">
En resumen, los cálculos renales representan una preocupación significativa para la salud humana debido a sus efectos negativos y a las limitaciones actuales en los métodos de diagnóstico. Sin embargo, la aplicación de redes neuronales en el análisis de orina ofrece la posibilidad de mejorar la precisión y eficiencia en el diagnóstico, lo que podría llevar a intervenciones tempranas y un mejor manejo de esta condición médica.

La utilidad de las RNAs no es solo el de una herramienta que puede llevar a cabo aquello de lo que ya es capaz el ser humano, sino que busca llevar esto más allá por medio de la capacidad de almacenamiento y procesamiento de información de las nuevas tecnologías computacionales. De este modo, ahí donde la limitada memoria y capacidad de procesamiento del ser humano le impide identificar patrones, clasificar o predecir, las RNAs pueden hacerlo.

## El perceptrón

<div style="text-align: justify"> 
Para que se pudiese crear una RNA primero fue necesario el desarrollo de la neurona artificial. La primera neurona de este tipo fue el <b>perceptrón</b></code></pre>. Esta fue desarrollada en los años 1950s y 1960s por Frank Rosenblatt. Una manera en la que podemos conceptualizar al perceptrón es como un dispositivo que toma decisiones a partir de alguna evidencia ponderada. En una versión muy simplificada del perceptrón, este recibe entradas binarias, efectua una ponderación de las mismss y finalmente emite una única salida tambíen binaria. 
  </div>

<div style="text-align: justify"> 
En la imagen, esta versión simplificada del perceptrón se encuentra recibiendo 3 entradas binarias: $x_1$, $x_2$ y $x_3$. ¿Cómo es que estas entradas son ponderadas? Rosenblatt propusó utilizar pesos $w$, en forma de números reales que determinen la relevancia que tiene cada entrada en la decisión final. En este caso, a cada una le correspondería un peso respectivo: $w_1$, $w_2$ y $w_3$. Finalmente, la decisión binaria final queda determinada por la siguiente regla: 
  </div>

\begin{eqnarray}
  \mbox{output} & = & \left\{ \begin{array}{ll}
      0 & \mbox{si } \sum_j w_j x_j \leq \mbox{ umbral} \\
      1 & \mbox{si } \sum_j w_j x_j > \mbox{ umbral}
      \end{array} \right.
\end{eqnarray}

<div style="text-align: justify">
Es decir, si la sumatoria de los productos entrada-peso ($w*x$) es mayor a un valor umbral predeterminado, entonces la salida será $1$. En cambio, si esta sumatoria es menor al umbral, entonces la salida será $0$. Podemos simplificar esta notación introduciendo el concepto de <b>sesgo</b></code></pre> del perceptrón. Este puede entenderse como la facilidad (o dificultad) que tiene un perceptrón dado para emitir un $1$ como respuesta. Su analogía directa con una célula neuronal es la facilidad con la que esta dispara dependiendo de sus propiedades biofísicas. En el caso del perceptrón, este sesgo puede representarse como un número real $b$, tal que entre más positivo, más fácil es el emitir un $1$; mientras que entre más negativo, más dificil se vuelve producir un $1$. Por lo tanto, nos damos cuenta que el sesgo puede plantearse como el opuesto del valor umbral ($b = -umbral$), ya que entre mayor es el umbral más difícil se vuelve para el perceptrón generar un $1$ como salida. Utilizando esta relación entre el sesgo y el umbral, y reescribiendo la sumatoria de los productos $w*x$ como un producto punto, podemos expresar los valores de posibles salida como:
</div>

\begin{eqnarray}
  \mbox{output} = \left\{ 
    \begin{array}{ll} 
      0 & \mbox{si } w\cdot x + b \leq 0 \\
      1 & \mbox{si } w\cdot x + b > 0
    \end{array}
  \right.
\end{eqnarray}

<div style="text-align: justify">
En otras palabras, si definimos a la variable $z = w\cdot x + b$, entonces la función $f(z) = 0$ cuando $z \leq 0$ y $f(z) = 1$ cuando $z > 0$. A esta función se le conoce como la <b>función de activación</b></code></pre> de la neurona artificial. Es decir, es la función que establece la relación entre las salidas de las neuronas, dependiendo de las entradas recibidas.   
  </div>

## La primera red neuronal artificial

<div style="text-align: justify"> 
Partiendo de la idea de que un perceptrón permite tomar una decisión simple a partir de evidencia simple, se vuelve plausible que un conjunto de perceptrones interconectados, nos permita procesar información y alcanzar decisiones más complejas. A este conjunto de perceptrones interconectados podemos adjudicarle ya el nombre de RNA. A continuacón se muestra una imagen de este tipo de RNA. 
</div>

<div style="text-align: justify"> 
Podemos darnos cuenta que la red está organizada en columnas o <b>capas</b></code></pre> de perceptrones. Aunque en la imagen se presenta, por cuestiones de simplificación, una red conformada por solo 3 capas, en realidad pueden existir muchas más. Notamos que la primera capa tiene un papel parecido al que tendría un perceptrón individual: recibir información simple y tomar decisiones simples; la segunda capa, al recibir el conjunto de decisiones de la primera, participa en la toma de decisiones a un nivel más complejo y abstracto. Como es de esperarse, en una tercera capa se llevarián a cabo decisiones aún más complejas. Eventualmente, toda esta información converge en una capa final normalmente conformada por un único perceptrón que emite la respuesta de la red. De este modo, una red conformada por muchas capas de perceptrones tiene la capacidad de efectuar decisiones sofisticadas. 
</div>

## El aprendizaje de la red y la neurona logística

<div style="text-align: justify">
Teniendo que cuenta que las salidas de una red de perceptrones solo pueden ser de tipo binario, son principalmente útiles en tareas de identificación. Este tipo de tarea consiste en que la red, a partir de la información provista, responda si la entrada particular pertenece o no a una categoría dada. Supongamos que estamos interesados en que nuestra RNA de perceptrones sea capaz de reconocer al número $"9"$ en una imagen digital. Digamos que esta imagen puede contener cualquier número del $0$ al $9$. En su estado inicial es de esperarse que su desempeño sea bastante bajo. Es decir, que muchas veces arroje respuestas positivas para números distintos de $9$ y respuestas negativas para imágenes que sí contienen al número $9$. No obstante, la red puede ser entrenada. Este entrenamiento consiste en evaluar su desempeño y con base en sus resultados modificar sus parámetros, como son los pesos y sesgos, a los cuales también se les conoce como <b>hiperparámetros</b></code></pre>. Esta actualización de los hiperparámetros se realiza con el objetivo de mejorar cada vez más el desempeño que tiene la red en la tarea de interés.
</div>

<div style="text-align: justify">
Aunque podría pensarse que a través del entrenamiento eventualmente se llegará al conjunto de valores que maximizan el desempeño de la red, al estar hablando de nodos binarios, existe un problema. Como se explicó antes, en estos nodos existe un punto umbral a partir del cual la neurona pasa de generar una salida a generar otra. Por lo tanto, es de esperarse que la red alcance un punto en el que incluso cambios sútiles en sus hiperparámetros hagan que la respuesta de algún perceptrón cambie, causando modificaciones impredecibles en la respuesta de la red.
</div>

<div style="text-align: justify">
La <b>neurona sigmoidal</b></code></pre> o <b>neurona logística</b></code></pre> surgió como solución a este problema. Aunque este tipo de neurona artificial puede plantearse de manera muy similar al perceptrón (recibe varias entradas a partir de las cuales genera una única salida), su principal diferencia radica en que su función de activación puede arrojar cualquier valor real entre $0$ y $1$. Esto permite que pequeñas modificaciones en los hiperparámetros de la neurona se manifiesten como pequeños cambios en sus sus salidas. Aunque en este caso la función de activación sigue dependiendo de $z = w⋅x+b$, ahora viene dada por $\sigma(z)$, expresada como:
</div>

\begin{eqnarray} 
  \sigma(z) \equiv \frac{1}{1+e^{-z}}.
\end{eqnarray}

Es decir:

\begin{eqnarray} 
  \frac{1}{1+\exp(-\sum_j w_j x_j-b)}.
\end{eqnarray}

<div style="text-align: justify">
A primera vista podría parecer algo muy distinto del perceptrón, sin embargo notamos que cuando $z$ es un número positivo grande, entonces $e^{-z} \approx 0$ y por lo tanto $\sigma(z) \approx 1$. Por otro lado, cuando $z$ es un número muy negativo, entonces $e^{-z} \rightarrow \infty$ y $\sigma(z) \approx 0$. En otras palabras, tal como ocurre en el perceptrón, valores de $z = w⋅x+b$ positivos grandes generan salidas $1$ y valores muy negativos generan salidas $0$.  
</div>

## Arquitectura de una red neuronal

<div style="text-align: justify">
Cada una de las capas que conforman a la red neuronal reciben un nombre particular. La primera capa, al estar conformada por las neuronas encargadas de recibir las entradas se le conoce como <b>capa de entradas</b></code></pre> (o <i>input layer</i></code></pre>). La última capa, al estar conformada por la o las neuronas encargadas de generar las salidas, se le conoce como <b>capa de salidas</b></code></pre> (o <i>output layer</i></code></pre>). Por último, a las capas que se encuentran entre estas dos se les conoce simplemente como <b>capas ocultas</b></code></pre> (o <i>hidden layers</i></code></pre>).
</div>

<div style="text-align: justify">
El diseño de las capas de entrada y salida es bastante directo. Si retomamos el ejemplo de la tarea de identificación del número $"9"$ en una imagen digital, podemos plantear que cada neurona de la capa de entrada reciba el nivel de intensidad de cada uno de los píxeles de la imagen. De este modo, la propia entrada determina las características de esta capa. Por otro lado, dado que como salida simplemente se busca una respuesta de tipo "sí/no" dependiendo de si la imagen pertenece a la categoría de interés, podemos diseñar esta capa con un único nodo logístico que arroje una respuesta positiva cuando se supera un umbral de $0.5$ o negativa cuando no se supera. El caso de las capas ocultas es distinto. No existe un método único para determinar la cantidad y estructura de estas capas. Por lo general esto se determina de manera empírica, sopesando por un lado el nivel de desempeño que alcanza la red, y por el otro el tiempo que toma el entrenamiento. 
</div> 

## Entrenamiento de una red neuronal

<div style="text-align: justify">
Continuemos con el ejemplo de una red para identificar si una imagen digital contiene o no el número $"9"$. Supongamos que utilizamos imagenes blanco y negro de $28x28$ píxeles. Esto implica que la red recibirá como entrada vectores de 784 dimensiones. A su vez, esto exige una capa de entrada conformada por 784 neuronas. Cada una de estas neuronas recibe el valor de intensidad en cada uno de los píxeles. Estos valores se encuentran en una escala del $0$ al $1$, donde cero representa un color blanco, $1$ representa un color negro, y los valores entre $0$ y $1$ representan las distintas escalas del color gris.  
</div>

<div style="text-align: justify">
Como mencionamos antes, la capa de salida está conformada por una única neurona logística arrojando valores entre $0$ y $1$. Mencionamos que cuando estos valores superan el umbral de $0.5$ entonces la red considera que la imagen presentada sí contiene al número $9$. En cambio, cuando el umbral no se supera, entonces la red arroja una respuesta negativa. Podemos considerar a estos valores de salida como la probabilidad de que la imagen contenga un $9$. Entre más cercano a $0$ sea este valor, menos probable es que, según la red, el número en la imagen sea un $9$. Entre más cercano a $1$ sea el valor, es más probable para la red que el número de la imagen sí sea un $9$. 
</div>

<div style="text-align: justify">
Si denotamos a los vectores de entrada de la red como $x$, entonces podemos usar la notación $y(x)$ para representar los valores esperados. Estos valores esperados son la probabilidad de que una imagen particular, descrita por el vector $x$, contenga al número $9$. Por lo tanto, estos valores pueden ser $0$ cuando la imagen contiene un número distinto de $9$ y $1$ cuando la imagen contiene al número $9$. Si denotamos el valor de salida de la red como $a$, entonces podemos comparar $a$ con $y(x)$ en cada ensayo para actualizar los pesos y sesgos, buscando que $a$ se aproxime cada vez más a $y(x)$. Para efectuar esta comparación necesitamos establecer una  <b>función de pérdida</b></code></pre>. Aunque existen muchas funciones de pérdida, una elección común, especialmente para tareas sencillas como la abordada aquí, es el <b>Error Medio Cuadrado</b></code></pre> o (EMC). Esta función queda expresada como sigue:
</div>

\begin{eqnarray}  C(w,b) \equiv
  \frac{1}{2n} \sum_x \| y(x) - a\|^2.
\end{eqnarray}

<div style="text-align: justify">
Aquí, $w$ denota al conjunto de todos los pesos de la red, $b$ al conjunto de todos los sesgos, $n$ es el total del numero de imagenes de entrenamiento, $a$ es la salida de la red cuando la entrada es $x$ y la sumatoria se realiza sobre todos los vectores de entrada $x$. La notación $∥v∥$ simplemente expresa la longitud de un vector $v$. Podemos notar que $C(w,b) \approx 0$ cuando $a \approx y(x)$, para todos los vectores $x$ de entrada. En cambio, los valores de $C(w,b)$ se vuelven más grandes a medida que la diferencia entre $a$ y $y(x)$ incrementa. Por lo tanto, en términos de la función de pérdida, el objetivo del entreamiento se convierte en encontrar el conjunto de pesos y sesgos que acercquen a $C(w,b)$ a $0$. Para esto necesitamos un <b>algoritmo de optimización</b></code></pre>.
</div>



<div style="text-align: justify">
Muchos de los algoritmos de optimización tienen como base el concepto de mínimo global provisto por el cálculo. Un ejemplo de este tipo de algoritmo es el de <b>descenso de gradiente</b></code></pre>. Como su nombre lo indica, este algoritmo se apoya del gradiente de nuestra función de pérdida. Como sabemos, el gradiente de una función se expresa con el símbolo $\nabla$ y consiste en un vector conformado por sus derivadas parciales y tiene la característica de que su dirección es la dirección de mayor crecimiento de la función. Por lo tanto, el vector gradiente del la función de pérdida queda expresado como:     
</div>

\begin{eqnarray} 
  \nabla C \equiv \left( \frac{\partial C}{\partial w}, 
  \frac{\partial C}{\partial b} \right).
\end{eqnarray}

<div style="text-align: justify">
Si la dirección de este vector es la de mayor crecimiento de la función, entonces su dirección opuesta $-\nabla C$ se vuelve la de mayor decrecimiento de la misma. Por lo tanto, el entrenamiento de nuestra red se resume en efectuar múltiples iteraciones en las que modifiquemos los valores de peso y sesgo, en la dirección contraria al gradiente. No obstante, surge la pregunta: ¿de qué tamaño deben ser las modificaciones de $b$ y $w$ en cada iteración? Esto es en realidad otro de los hiperparámetros de la red y se conoce como <b>tasa de aprendizaje</b></code></pre> ($\eta$). Este hiperparámetro también debe de ajustarse probando distintos valores en distintas sesiones de entrenamiento y sopesando el desempeño alcanzado por la red con la duración del entrenamiento. 
    </div>

Una vez seleccionado el valor de $\eta$, si denotamos como $v = (w,b)$, entonces el cambio en los valores de $v$ en cada iteración queda expresado como: 

\begin{eqnarray} 
  \Delta v = -\eta \nabla C
\end{eqnarray}

El final de una sesión de entrenamiento está determinado por un valor umbral en términos de la función de pérdida. Es decir, el algoritmo de optimización seguirá modificando los valores peso y sesgo, hasta que la función de pérdida se vuelva menor a este valor umbral pre-establecido. Cabe mencionar que este valor umbral es también un hiperparámetro que puede ajustarse a lo largo de múltiples sesiones de entrenamiento. 

El resultado esperado consiste en una red capaz de resolver nuestra tarea de interés con una alta tasa de éxito. En este caso, se esperaría que tras haber ajustado los distintos parámetros, la red sea capaz de reconocer cuando una imagen contiene o no al número $9$ con solo un pequeño margen de error.

## Proyecto: Diágnostico de Cálculos Renales Mediante Análisis de Orina Utilizando Redes Neuronales

### Introducción

<p align="justify">
Los cálculos renales, también conocidos como litiasis renal o nefrolitiasis, son una afección común que afecta a un gran número de personas en todo el mundo. Estos son pequeños depósitos sólidos que se forman en los riñones a partir de sustancias presentes en la orina. Los cálculos renales pueden variar en tamaño, desde diminutos granos de arena hasta piedras más grandes que pueden obstruir los conductos urinarios.

<p align="justify">
Existen diferentes tipos de cálculos renales, y su composición puede estar compuesta principalmente de sales de calcio, ácido úrico, estruvita o cistina. Las causas exactas de su formación pueden variar, pero en general, los factores de riesgo incluyen una combinación de desequilibrios químicos en la orina, baja ingesta de líquidos, dieta poco saludable, obesidad, historial familiar de cálculos renales y ciertas condiciones médicas como la hiperparatiroidismo o la enfermedad inflamatoria intestinal.

<p align="justify">
Los efectos negativos de los cálculos renales en la salud pueden ser significativos. Estos pueden causar dolor intenso en el costado o la espalda baja, así como síntomas como náuseas, vómitos, sangre en la orina e infecciones recurrentes del tracto urinario. Además, los cálculos renales pueden bloquear el flujo normal de orina, lo que puede llevar a complicaciones graves, como daño renal o infecciones graves.

<p align="justify">
El diagnóstico de los cálculos renales se realiza generalmente mediante una combinación de síntomas clínicos, análisis de orina, pruebas de imagen y, en algunos casos, pruebas de laboratorio adicionales. Sin embargo, los métodos de diagnóstico convencionales pueden presentar limitaciones. Por ejemplo, los análisis de orina actuales pueden ser subjetivos y dependen de la interpretación del especialista. Además, las pruebas de imagen pueden no detectar cálculos pequeños o pueden requerir radiación ionizante, lo que plantea preocupaciones en términos de exposición y seguridad.

<p align="justify">
En este contexto, las redes neuronales, un enfoque de inteligencia artificial inspirado en el funcionamiento del cerebro humano, han surgido como una herramienta prometedora en el diagnóstico de cálculos renales. Estas redes neuronales pueden aprender patrones complejos a partir de grandes conjuntos de datos, lo que permite desarrollar modelos capaces de detectar y clasificar cálculos renales en base al análisis de orina.

<p align="justify">
Las estadísticas revelan que los cálculos renales son una afección común en todo el mundo, afectando aproximadamente al 10% de la población en algún momento de su vida. Además, la incidencia de cálculos renales parece estar en aumento, posiblemente debido a cambios en la dieta y los estilos de vida.

<p align="justify">
En resumen, los cálculos renales representan una preocupación significativa para la salud humana debido a sus efectos negativos y a las limitaciones actuales en los métodos de diagnóstico. Sin embargo, la aplicación de redes neuronales en el análisis de orina ofrece la posibilidad de mejorar la precisión y eficiencia en el diagnóstico, lo que podría llevar a intervenciones tempranas y un mejor manejo de esta condición médica.
