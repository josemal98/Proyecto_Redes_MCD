## El perceptrn

<div style="text-align: justify"> 
Para que se pudiese crear una RNA primero fue necesario el desarrollo de la neurona artificial. La primera neurona de este tipo fue el <b>perceptrn</b></code></pre>. Esta fue desarrollada en los aos 1950s y 1960s por Frank Rosenblatt. Una manera en la que podemos conceptualizar al perceptrn es como un dispositivo que toma decisiones a partir de alguna evidencia ponderada. En una versin muy simplificada del perceptrn, este recibe entradas binarias, efectua una ponderacin de las mismss y finalmente emite una nica salida tamben binaria.

<div style="text-align: justify"> 
En la imagen, esta versin simplificada del perceptrn se encuentra recibiendo 3 entradas binarias: $x_1$, $x_2$ y $x_3$. Cmo es que estas entradas son ponderadas? Rosenblatt propus utilizar pesos $w$, en forma de nmeros reales que determinen la relevancia que tiene cada entrada en la decisin final. En este caso, a cada una le correspondera un peso respectivo: $w_1$, $w_2$ y $w_3$. Finalmente, la decisin binaria final queda determinada por la siguiente regla: 
  </div>

\begin{eqnarray}

      0 & \mbox{si } \sum_j w_j x_j \leq \mbox{ umbral} \\
      1 & \mbox{si } \sum_j w_j x_j > \mbox{ umbral}
      \end{array} \right.
\end{eqnarray}

<div style="text-align: justify">
Es decir, si la sumatoria de los productos entrada-peso ($w*x$) es mayor a un valor umbral predeterminado, entonces la salida ser $1$. En cambio, si esta sumatoria es menor al umbral, entonces la salida ser $0$. Podemos simplificar esta notacin introduciendo el concepto de <b>sesgo</b></code></pre> del perceptrn. Este puede entenderse como la facilidad (o dificultad) que tiene un perceptrn dado para emitir un $1$ como respuesta. Su analoga directa con una clula neuronal es la facilidad con la que esta dispara dependiendo de sus propiedades biofsicas. En el caso del perceptrn, este sesgo puede representarse como un nmero real $b$, tal que entre ms positivo, ms fcil es el emitir un $1$; mientras que entre ms negativo, ms dificil se vuelve producir un $1$. Por lo tanto, nos damos cuenta que el sesgo puede plantearse como el opuesto del valor umbral ($b = -umbral$), ya que entre mayor es el umbral ms difcil se vuelve para el perceptrn generar un $1$ como salida. Utilizando esta relacin entre el sesgo y el umbral, y reescribiendo la sumatoria de los productos $w*x$ como un producto punto, podemos expresar los valores de posibles salida como:
</div>

\begin{eqnarray}
  \mbox{output} = \left\{ 
    \begin{array}{ll} 
      0 & \mbox{si } w\cdot x + b \leq 0 \\
      1 & \mbox{si } w\cdot x + b > 0
    \end{array}
  \right.
\end{eqnarray}

<div style="text-align: justify">
En otras palabras, si definimos a la variable $z = w\cdot x + b$, entonces la funcin $f(z) = 0$ cuando $z \leq 0$ y $f(z) = 1$ cuando $z > 0$. A esta funcin se le conoce como la <b>funcin de activacin</b></code></pre> de la neurona artificial. ## El aprendizaje de la red y la neurona logstica

<div style="text-align: justify">
Teniendo que cuenta que las salidas de una red de perceptrones solo pueden ser de tipo binario, son principalmente tiles en tareas de identificacin. <div style="text-align: justify">
Aunque podra pensarse que a travs del entrenamiento eventualmente se llegar al conjunto de valores que maximizan el desempeo de la red, al estar hablando de nodos binarios, existe un problema. <div style="text-align: justify">
La <b>neurona sigmoidal</b></code></pre> o <b>neurona logstica</b></code></pre> surgi como solucin a este problema. Aunque este tipo de neurona artificial puede plantearse de manera muy similar al perceptrn (recibe varias entradas a partir de las cuales genera una nica salida), su principal diferencia radica en que su funcin de activacin puede arrojar cualquier valor real entre $0$ y $1$. \begin{eqnarray}

\end{eqnarray}

\begin{eqnarray} 
  \frac{1}{1+\exp(-\sum_j w_j x_j-b)}.

A primera vista podra parecer algo muy distinto del perceptrn, sin embargo notamos que cuando $z$ es un nmero positivo grande, entonces $e^{-z} \approx 0$ y por lo tanto $\sigma(z) \approx 1$. ## Entrenamiento de una red neuronal

<div style="text-align: justify">
Continuemos con el ejemplo de una red para identificar si una imagen digital contiene o no el nmero $"9"$. Supongamos que utilizamos imagenes blanco y negro de $28x28$ pxeles. <div style="text-align: justify">
Como mencionamos antes, la capa de salida est conformada por una nica neurona logstica arrojando valores entre $0$ y $1$. </div>

<div style="text-align: justify">
Si denotamos a los vectores de entrada de la red como $x$, entonces podemos usar la notacin $y(x)$ para representar los valores esperados. Si denotamos el valor de salida de la red como $a$, entonces podemos comparar $a$ con $y(x)$ en cada ensayo para actualizar los pesos y sesgos, buscando que $a$ se aproxime cada vez ms a $y(x)$. Para efectuar esta comparacin necesitamos establecer una  <b>funcin de prdida</b></code></pre>. Aunque existen muchas funciones de prdida, una eleccin comn, especialmente para tareas sencillas como la abordada aqu, es el <b>Error Medio Cuadrado</b></code></pre> o (EMC). Esta funcin queda expresada como sigue:
</div>

\begin{eqnarray}  C(w,b) \equiv
  \frac{1}{2n} \sum_x \| y(x) - a\|^2.
\end{eqnarray}

<div style="text-align: justify">
Aqu, $w$ denota al conjunto de todos los pesos de la red, $b$ al conjunto de todos los sesgos, $n$ es el total del numero de imagenes de entrenamiento, $a$ es la salida de la red cuando la entrada es $x$ y la sumatoria se realiza sobre todos los vectores de entrada $x$. La notacin $v$ simplemente expresa la longitud de un vector $v$. Podemos notar que $C(w,b) \approx 0$ cuando $a \approx y(x)$, para todos los vectores $x$ de entrada. En cambio, los valores de $C(w,b)$ se vuelven ms grandes a medida que la diferencia entre $a$ y $y(x)$ incrementa. Por lo tanto, en trminos de la funcin de prdida, el objetivo del entreamiento se convierte en encontrar el conjunto de pesos y sesgos que acercquen a $C(w,b)$ a $0$. Para esto necesitamos un <b>algoritmo de optimizacin</b></code></pre>.
</div>



<div style="text-align: justify">
Muchos de los algoritmos de optimizacin tienen como base el concepto de mnimo global provisto por el clculo. Un ejemplo de este tipo de algoritmo es el de <b>descenso de gradiente</b></code></pre>. Como su nombre lo indica, este algoritmo se apoya del gradiente de nuestra funcin de prdida. Como sabemos, el gradiente de una funcin se expresa con el smbolo $\nabla$ y consiste en un vector conformado por sus derivadas parciales y tiene la caracterstica de que su direccin es la direccin de mayor crecimiento de la funcin. Por lo tanto, el vector gradiente del la funcin de prdida queda expresado como:     
</div>
